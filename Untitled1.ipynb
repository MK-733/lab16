```python
import numpy as np
import matplotlib.pyplot as plt

Is = 1e-12
n = 1.8
Vt = 0.026

V = np.linspace(0, 1.0, 500)

I = Is * (np.exp(V / (n * Vt)) - 1)

I[I > 0.06] = 0.06

plt.figure(figsize=(8, 5))
plt.plot(V, I)

plt.xlabel("voltage, Volts")
plt.ylabel("current, Amps")
plt.title("Iâ€“V characteristic")
plt.grid(True)

plt.show()
```
