```python
import numpy as np
import matplotlib.pyplot as plt

Is = 1e-9
n = 1.8
Vt = 0.031

V = np.linspace(0, 1.0, 900)

I = Is * (np.exp(V / (n * Vt)) - 1)

I[I > 0.06] = 0.06

plt.figure(figsize=(7,5))
plt.plot(V, I)

plt.xlabel("voltage, Volts")
plt.ylabel("current, Amps")
plt.title("I-V characteristic")
plt.grid(True)

plt.show()
```
